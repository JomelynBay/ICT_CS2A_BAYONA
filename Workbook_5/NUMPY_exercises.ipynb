A.

```python
import numpy as np  # Importing numpy library

# Creating a nested Python list with values from 1 to 100
nested_list = [[i for i in range(1, 101)]]

# Converting the nested list into a numpy ndarray object
A = np.array(nested_list)

# Printing the shape of the ndarray object
print("Shape of A:", A.shape)

# Finding the maximum value in the ndarray object using a method
print("Maximum value in A:", A.max())

# Using a numpy function to find the maximum value in A
print("Maximum value in A (using np.max):", np.max(A))
```

B.

```python
# Creating a 5x5 integer array filled with ones
ones_array = np.ones((5, 5), dtype=int)

# Printing the shape, size, and number of dimensions of the array
print("Shape of ones_array:", ones_array.shape)
print("Size of ones_array:", ones_array.size)
print("Number of dimensions of ones_array:", ones_array.ndim)
```

A. Create a four-dimensional array of random integers in the interval [0, 12]. Print the result.

Get the dimensions, shape, and size.
Explain what is the result briefly by writing a comment before the code.

C. Multipy each array with 4. What is its result? * Write a brief explanation on the multiplication process.

```python
# Creating a four-dimensional array of random integers in the interval [0, 12]
import numpy as np
four_dim_array = np.random.randint(0, 13, size=(3, 4, 5, 6))

# Printing the array
print("Four-dimensional array:\n", four_dim_array)

# Getting dimensions, shape, and size
print("Dimensions:", four_dim_array.ndim)
print("Shape:", four_dim_array.shape)
print("Size:", four_dim_array.size)

# Getting the maximum number for each dimension and displaying it
max_values = [four_dim_array[:, :, :, i].max() for i in range(four_dim_array.shape[3])]
print("Maximum number for each dimension:", max_values)

# Multiplying each array element by 4
multiplied_array = four_dim_array * 4

# Printing the result
print("Result after multiplying each array element by 4:\n", multiplied_array)

#Every integer value in the four-dimensional array is multiplied by four when each array element is multiplied by four.
#Throughout the entire array, this procedure is applied element by element.
#Each element in the resultant array is multiplied by four, and it has the same shape as the original array.
```
