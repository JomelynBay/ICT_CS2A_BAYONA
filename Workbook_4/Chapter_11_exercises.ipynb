```python
%%writefile my_module.py
# my_module.py

"""
A simple module variables and functions
"""

# Define a variable
my_var = "Hello from my_module!"

# Define a function
def greet(name):
    """
    Greet the user.

    Parameters:
        name (str): The name of the user.

    Returns:
        str: A greeting message.
    """
    greeting = f"Hello, {name}! Welcome to my_module."
    return greeting
```

```python

import my_module

# Access the variable
print(my_module.my_var)

# Call the function
greeting_message = my_module.greet("Kathleen")
print(greeting_message)
```
