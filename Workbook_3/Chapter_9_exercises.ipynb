Exercise 3
Write a program that asks the user to enter a line of text and then counts the number of each word that appeared in the line of text. Use the string function split() to make a list of words from a given string e.g.:

word_list = line.split()
will return a list of words from the string line and store it in the variable word_list.

```python
# Ask the user to enter a line of text
line = input("Enter a line of text: ")

# Split the line into a list of words
word_list = line.split()

# Create an empty dictionary to store word counts
word_counts = {}

# Iterate through each word in the word list
for word in word_list:
    # Increment the count for the word in the dictionary
    word_counts[word] = word_counts.get(word, 0) + 1

# Display the word counts
print("Word Counts:")
for word, count in word_counts.items():
    print(f"{word}: {count}")
```
