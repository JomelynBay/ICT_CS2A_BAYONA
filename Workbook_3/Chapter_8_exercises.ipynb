#### Exercise 3

Create a list containing the valid grade values, then using `enumerate` in a `for` loop, print out all **indices** and **items** in the list.

It should print out:

    0   1.0
    1   1.25
    2   1.5
    3   1.75
    4   2.0
    5   2.25
    6   2.5
    7   3.0
      

```python
my_grades = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]
for index, grade in enumerate(my_grades):
  print(index, grade, sep="\t")
```

#### Exercise 4

```python
higher_than = [grade for grade in my_grades if grade > 2.0]
higher_than
```

## CHALLENGE

Modify the code in the subheading [Parsing Lines](#ParsingLines) to display all the time stamps
of the emails in mboxt-short.txt


```python
fhand = open('mbox-short.txt')
my_list = []
for line in fhand:
    line = line.rstrip()
    if not line.startswith('From'):
        continue
    words = line.split()
    my_list.append(" ".join(words[2:]))
print(my_list)
```
