EXERCISE 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
def compute_pay(hours,rate):
    if hours <= 40:
        pay = hours * rate
    else:
        pay = 40 * rate +(1.5 * (hours - 40)*rate)
    return pay

try:
  hours=float(input("enter hours: "))
  rate=float(input("enter rate: "))
  pay=compute_pay(hours,rate)
  print("pay: ",pay)
except:
  print("invalid input")
```

CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
score=int(input('Enter the score: '))

if score >= 97:
  print('Grade: 1.0')
elif 93 <= score <= 97:
  print('Grade: 1.25')
elif 89 <= score <= 93:
  print('Grade: 1.5')
elif 85 <= score <= 89:
  print('Grade: 1.75')
elif 81 <= score <= 85:
  print('Grade: 2.0')
elif 77 <= score <= 81:
  print('Grade: 2.25')
elif 74 <= score <= 77:
  print('Grade: 2.5')
elif 0 <= score <= 74:
  print('Grade: 3.0')
```
